```python
from funciones import es_paquete_valido, clasificar_porcentaje

# ---------------- PROGRAMA PRINCIPAL -----------------
paquetes_correctos = []
# a) Pedir 3 paquetes
for i in range(1, 4):  #1234567899
    paquete = input(f"Introduce el paquete {i} (10 piezas entre 1-4): ")
    
    # b) Validar
    if es_paquete_valido(paquete):
        print(f"✔ Paquete {i} válido: {paquete}")
        paquetes_correctos.append(paquete)
    else:
        print(f"❌ Paquete {i} incorrecto y descartado")
print("\n----- RESULTADOS -----")
# Pesos
pesos = {
    "1": 0.5,
    "2": 3,
    "3": 5,
    "4": 0.2
}

for paquete in paquetes_correctos:  #[1324567899, 2233441122, 1111111111]
    
    # Contar piezas
    n2 = paquete.count("2")
    n3 = paquete.count("3")
    
    porcentaje = ((n2 + n3) / 10) * 100
    
    # Clasificación
    clasificacion = clasificar_porcentaje(porcentaje)
    
    # Calcular peso total
    peso_total = sum(pesos[p] for p in paquete) #Recorre cada carácter del paquete como p . Primero piensa en for p in paquete
                                                #entonces ese p, lo pasamos al diccionario pesos
                                                #y me dara el peso correspondiente. Sum lo suma todo.

    # Mostrar resultados. OJO esto esta dentro del bucle
    print("\nPaquete:", paquete)
    print(f"Piezas tipo 2: {n2}, tipo 3: {n3}")
    print(f"Porcentaje tipo2+tipo3: {porcentaje:.2f}%")
    print(f"Peso total: {peso_total:.2f} kg")
    print(f"Clasificación: {clasificacion}")
```
